```python
import pandas as pd
```

```python
df_train = pd.read_csv("train.csv")
```

```python
df_train["date"] = pd.to_datetime(df_train["date"])
```

```python
df_train["day_of_week"] = df_train["date"].dt.day_name()
```

```python
df_train["date_of_year"] = df_train["date"].dt.strftime('%m-%d')
```

```python
df_train.rename(columns={"date_of_year":"day_of_year"},inplace=True)
```
