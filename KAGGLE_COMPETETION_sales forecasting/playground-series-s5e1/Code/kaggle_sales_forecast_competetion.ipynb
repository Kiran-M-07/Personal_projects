```python
import pandas as pd
import matplotlib.pyplot as plt
from prophet import Prophet

# Load training and test data
train_data = data2.copy()  # Replace with actual file path
# test_data = pd.read_csv("your_test_data.csv")  # Replace with actual file path

# Convert date to datetime format
train_data["date"] = pd.to_datetime(train_data["date"])
test_data["date"] = pd.to_datetime(test_data["date"])

# Prepare test dataset format
test_data["num_sold"] = None  # Placeholder for predictions

# Function to train Prophet and predict on test data
def forecast_sales(group, test_group):
    # Rename columns for Prophet
    prophet_df = group.rename(columns={"date": "ds", "num_sold": "y"})
    
    # Initialize and train Prophet model
    model = Prophet()
    model.fit(prophet_df)
    
    # Use test dates for forecasting
    future = test_group[["date"]].rename(columns={"date": "ds"})
    forecast = model.predict(future)
    
    # Merge predictions with test data
    test_group["predicted_num_sold"] = forecast["yhat"].values
    test_group["yhat_lower"] = forecast["yhat_lower"].values
    test_group["yhat_upper"] = forecast["yhat_upper"].values
    
    return test_group

# Apply forecasting for each (country, store, product) in test data
predictions = []
grouped_train = train_data.groupby(["country", "store", "product"])
grouped_test = test_data.groupby(["country", "store", "product"])

for group_key, train_group in grouped_train:
    if group_key in grouped_test.groups:
        test_group = grouped_test.get_group(group_key)
        pred = forecast_sales(train_group, test_group)
        predictions.append(pred)

# Combine all predictions
predictions_df = pd.concat(predictions)

# Save results
predictions_df.to_csv("test_predictions.csv", index=False)
print(predictions_df.head())  # Display first few rows
```
